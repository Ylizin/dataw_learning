```python
tips = pd.read_csv('examples/tips.csv')
# Add tip percentage of total bill
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips[:6]
```

```python
grouped = tips.groupby(['day', 'smoker'])
```

```python
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
grouped_pct.agg(['mean', 'std', peak_to_peak]) #统计出三列
```

```python
grouped_pct.agg([('foo', 'mean'), ('bar', np.std)]) #为mean/std等列附上名称
```

```python
functions = ['count', 'mean', 'max']
result = grouped['tip_pct', 'total_bill'].agg(functions)
result
```

```python
result['tip_pct']
```

```python
ftuples = [('Durchschnitt', 'mean'), ('Abweichung', np.var)]
grouped['tip_pct', 'total_bill'].agg(ftuples) #此处grouped的索引拥有最高的level
```

```python
grouped.agg({'tip' : np.max, 'size' : 'sum'})
```

```python
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
             'size' : 'sum'}) #指定层次
```

```python
tips.groupby(['day', 'smoker'], as_index=False).mean() #当as index为false的时后，index就会被解开
```

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
result.unstack('smoker')
```

```python
tips.groupby('smoker', group_keys=False).apply(top)
```

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
                      'data2': np.random.randn(1000)})
quartiles = pd.cut(frame.data1, 4) #cut可以得到数据对应的桶
quartiles[:10]
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}
grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats).unstack()
```

```python
states = ['Ohio', 'New York', 'Vermont', 'Florida',
          'Oregon', 'Nevada', 'California', 'Idaho']
group_key = ['East'] * 4 + ['West'] * 4
data = pd.Series(np.random.randn(8), index=states)
data
```

```python
data[['Vermont', 'Nevada', 'Idaho']] = np.nan
data
```

```python
data.groupby(group_key).mean()
```

```python
fill_mean = lambda g: g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
fill_values = {'East': 0.5, 'West': -1}
fill_func = lambda g: g.fillna(fill_values[g.name])
data.groupby(group_key).apply(fill_func)
```

```python
# Hearts, Spades, Clubs, Diamonds
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)
```

```python
deck[:13]
```

```python
def draw(deck, n=5):
    return deck.sample(n)
draw(deck)
```

```python
get_suit = lambda card: card[-1] # last letter is suit
deck.groupby(get_suit).apply(draw, n=2)
```

```python
deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

```python
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a',
                                'b', 'b', 'b', 'b'],
                   'data': np.random.randn(8),
                   'weights': np.random.rand(8)})
df
```

```python
close_px = pd.read_csv('examples/stock_px_2.csv', parse_dates=True,
                       index_col=0)
close_px.info()
```

```python
close_px[-4:]
```

```python
spx_corr = lambda x: x.corrwith(x['SPX'])
```

```python
rets = close_px.pct_change().dropna()
```

```python
get_year = lambda x: x.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda g: g['AAPL'].corr(g['MSFT']))
```

```python
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, 'AAPL', ['SPX'])
```
