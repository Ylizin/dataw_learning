```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.arange(0,3*np.pi,0.01)
C = np.cos(data)
S = np.sin(data)
```

```python
plt.subplot(2,1,1)
plt.title('sine')
plt.plot(data,S)
plt.subplot(2,1,2)
plt.title('cosine')
plt.plot(data,C)
```

```python
plt.title('sine and sine')
plt.plot(data,S)
plt.plot(data,C)
```
